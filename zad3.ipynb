```python

import numpy as np
import matplotlib.pyplot as plt
import skimage.transform
from PIL import Image

import skimage.color
```

```python
im = np.array(Image.open('zad3_1.jpg'))
plt.figure(figsize=(8,8))

plt.imshow(im)
```

```python
im2 = np.array(Image.open('zad3_2.jpg'))
plt.figure(figsize=(8,8))

plt.imshow(im2)
```

```python
im3 = im + im2
im3 = im3/255.0

im3 = (im3 - np.min(im3))/(np.max(im3) - np.min(im3))
plt.figure(figsize=(8,8))
plt.imshow(im3)
plt.imshow(np.mean(im3, axis=2)**0.6)
plt.colorbar()
plt.clim(0, 0.05)
plt.imshow(im3**0.33)
```

```python

im1_hsv = skimage.color.rgb2hsv(im3)

plt.imshow(im1_hsv[:,:,1])
plt.colorbar()
mask = im1_hsv[:,:,1] < 0.1
plt.imshow(mask)
plt.colorbar()
ind = np.argwhere(mask==1)
rows = ind[:,0]
cols = ind[:,1]
white_pixels1 = im[rows, cols, :]
white_pixels1.shape
white1 = np.mean(white_pixels1, axis = 0)
white1.shape

white_pixels2 = im2[rows, cols, :]
white2 = np.mean(white_pixels2, axis = 0)
white2

white_pixels3 = im3[rows, cols, :]
white3 = np.mean(white_pixels3, axis = 0)
white3

corr_factors_2 = white1/white2
corr_factors_2

corr_factors_3 = white1/white3
corr_factors_3

plt.imshow(im3*corr_factors_2)
```
