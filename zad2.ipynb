```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

im = np.array(Image.open('zad2_1.jpg'))
```

```python
  im_g = np.mean(im, axis = 2)/255.0
```

```python
im_high = im_g * 1.4

plt.figure(figsize=(12, 6))

plt.subplot(1,2,2)
plt.imshow(im_high, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Zwiększony kontrast')
```

```python

im2 = np.array(Image.open('zad2_2.jpg'))
plt.figure(figsize = (8, 8))

im_p = np.mean(im2, axis = 2)/255.0

im2_05 = im_p**0.5

plt.figure(figsize=(12, 6))

plt.subplot(1,2,2)
plt.imshow(im2_05, cmap='gray')
plt.clim(0, 1)
plt.colorbar()
plt.title('Korekcja gamma')
```
